#### Вторая модель логистической регрессии

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop
```

```python
def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]
```

#### Третья модель логистической регрессией с регуляризацией

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Гиперпараметры
C = [0.1, 0.01]
iters = [50, 100, 200]
epsilon_stop = [0.0001]
param_grid = [
    {'C': C,
     'penalty': ['l1'],
     'solver': ['saga', 'liblinear'],
     'class_weight':[None, 'balanced'],
     'multi_class': ['auto', 'ovr'],
     'max_iter': iters,
     'tol': epsilon_stop},
    {'C': C,
     'penalty': ['l2'],
     'solver': ['newton-cg', 'liblinear', 'lbfgs', 'saga'],
     'class_weight':[None, 'balanced'],
     'multi_class': ['auto', 'ovr'],
     'max_iter': iters,
     'tol':epsilon_stop},
    {'penalty': ['none'],
     'solver': ['saga'],
     'class_weight':[None, 'balanced'],
     'multi_class': ['auto', 'ovr'],
     'max_iter': iters,
     'tol':epsilon_stop}]
```

```python
def fit_logreg_oversampling_grid(model, param_grid=param_grid):
    X = model.iloc[:,:-1]
    y = model.iloc[:,-1]
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.20, random_state=42)
    clf = LogisticRegression()
    oversam = RandomOverSampler()
    X_over, y_over = oversam.fit_resample(X_train, y_train)
    clf.fit(X_over, y_over)
    
    # Произведем поиск гиперпараметров GridSearchCV при помощи перебора параметров по сетке param_grid:
    grid_search = GridSearchCV(clf, param_grid, scoring='recall', n_jobs=-1)
    grid_search.fit(X_over, y_over)
    grid_search.best_params_
    
    # Печатаем параметры развернуто:
    best_model = grid_search.best_estimator_
    best_parameters = best_model.get_params()
    for param_name in sorted(best_parameters.keys()):
        print('\t%s: %r' % (param_name, best_parameters[param_name]))
    
    y_pred = best_model.predict(X_valid)
        
    print('Recall: ', recall_score(y_valid, y_pred, average='macro'))
    print('f1_score: ', f1_score(y_valid, y_pred))
```

```python
fit_logreg_oversampling_grid(arterial_model_2, param_grid)
```
